```python
import pandas as pd

# Cargar el dataset
df = pd.read_csv("hotel_bookings.csv")

# Mostrar las primeras 5 filas
df.head()
df.info() 
```

```python
# mostrar todas las columnas
df.columns
```

```python
# mostrar todas las columnas del dataframe
for col in df.columns:
    print(col)
```

```python
# verificar si hay valores nulos
df.isnull().sum()
```

```python
# Remplazar valores nulos de children con 0
df['children'].fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
# quitar valores nulos de country con Unknown
df['country'].fillna('Unknown', inplace=True)

df.isnull().sum()
```

```python
# quitar valores nulos de agent con 0 y company con 0
df['agent'].fillna(0, inplace=True)
df['company'].fillna(0, inplace=True)

df.isnull().sum()

```

```python
# guardar el dataframe limpio en un csv
df.to_csv('tutoria.csv', index=False)

# cargar el dataframe limpio
df = pd.read_csv('tutoria.csv')
```

```python
# Contar las reservas por el tipo hotel
hotel_count = df['hotel'].value_counts()
print(hotel_count)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Graficar la cantidad de reservas por el tipo hotel
sns.barplot(x=hotel_count.index, y=hotel_count.values)
plt.title('Cantidad de reservas por tipo hotel')
plt.xlabel('Tipo hotel')
plt.ylabel('Cantidad de reservas')
plt.show()
```

```python
# Calcular la tasa de cancelacion
cancellation_rate = df['is_canceled'].mean()
print(f'Tasa de cancelacion: {cancellation_rate:.2%}')
```

```python
# ingresos promedio por tipo de hotel
average_adr = df.groupby('hotel')['adr'].mean()
print(average_adr)

print('Ingresos Promedio por tipo de hotel: ')
for hotel, avg_adr in average_adr.items():
    print(f'{hotel}: {avg_adr:.2f}$')
```

```python
# Tasa de cancelacion por tipo de cliente
cancellation_rate = df.groupby('customer_type')['is_canceled'].mean()
# print(cancellation_rate)

print('Tasa de cancelacion por tipo de cliente: ')
for customer_type, rate in cancellation_rate.items():
    print(f'{customer_type}: {rate:.2%}')
```

```python
# Crear una columna de fechas   
df['arrive_date']= pd.to_datetime(df['arrival_date_year'].astype(str)+'-'+df['arrival_date_month'].astype(str)+'-'+df['arrival_date_day_of_month'].astype(str))

# contar reservas por mes
monthly_counts = df['arrive_date'].dt.to_period('M').value_counts().sort_index()
print(monthly_counts)
```

```python
monthly_counts.plot(kind='line')
plt.title('Reservas por Mes')
plt.xlabel('Mes')
plt.ylabel('Cantidad de Reservas')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Supongamos que monthly_counts es tu serie de tiempo con el ndice como meses
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Eje X: los meses (como nmeros)
x = np.arange(len(monthly_counts))
# Eje Y: cantidad de reservas
y = monthly_counts.values
# Eje Z: puedes poner todo en cero o usar el ndice para ver la evolucin
z = np.zeros_like(x)

ax.plot(x, y, z, label='Reservas por Mes')
ax.set_title('Reservas por Mes en 3D')
ax.set_xlabel('Mes')
ax.set_ylabel('Cantidad de Reservas')
ax.set_zlabel('Evolucin')
ax.legend()
plt.show()
```
